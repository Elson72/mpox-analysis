```python
total_cases = mpox_df.groupby(['year', 'country'])['new_confirmed_cases'].sum().reset_index()
```

```python
total_cases
```

```python
total_deaths = mpox_df.groupby(['year', 'country'])['new_deaths'].sum().reset_index()
```

```python
total_deaths
```

```python
pivot_total_cases = total_cases.pivot(index='country', columns='year', values='new_confirmed_cases').fillna(0)
```

```python
pivot_total_cases.plot(kind='bar', stacked=True, figsize=(10, 6))

sns.set(style="whitegrid")
colors = sns.color_palette("deep", n_colors=len(pivot_total_cases.columns))
ax = pivot_total_cases.plot(kind='bar', stacked=True, figsize=(10, 6), color=colors)
```

```python
plt.title('Total Cases by Year and Country', fontsize=16 )
plt.xlabel('country', fontsize=12)
plt.ylabel('Total Cases', fontsize=12)
plt.legend(title='year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
pivot_total_deaths = total_deaths.pivot(index='country', columns='year', values='new_deaths').fillna(0)
pivot_total_deaths.plot(kind='bar', stacked=True, figsize=(12, 8))

sns.set(style="whitegrid")
plt.clf()
colors = sns.color_palette("deep", n_colors=len(pivot_total_cases.columns))
ax = pivot_total_cases.plot(kind='bar', stacked=True, figsize=(12, 8), color=colors)

plt.title('Total Deaths by Year and Country', fontsize=16 )
plt.xlabel('country')
plt.ylabel('Total Deaths', fontsize=12)
y_ticks = range(0, int(pivot_total_deaths.values.sum().max()) + 10, 10)  
plt.yticks(y_ticks)
plt.legend(title='year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
mpox_df['CFR (%)'] = mpox_df.apply(
    lambda row: (row['new_deaths'] / row['new_confirmed_cases']) * 100 if row['new_confirmed_cases'] > 0 else 0,
    axis=1
)
```

```python
country_cfr = mpox_df.groupby('country', as_index=False).agg(
    Total_Confirmed_Cases=('new_confirmed_cases', 'sum'),
    Total_Deaths=('new_deaths', 'sum'),
    CFR=('CFR (%)', 'mean')
)

print(country_cfr[['country', 'Total_Confirmed_Cases', 'Total_Deaths', 'CFR']])
```

```python
top_10_countries = total_cases.nlargest(10, 'new_confirmed_cases')['country'].values

# Filter the original dataset for the top 10 countries
```

```python
top_10_countries
```

```python
top_10_data = mpox_df[mpox_df['country'].isin(top_10_countries)]
```

```python
top_10_data
```

```python
top_10_data['z_score'] = (top_10_data['new_confirmed_cases'] - top_10_data['new_confirmed_cases'].mean()) / top_10_data['new_confirmed_cases'].std()
top_10_data['is_outlier'] = np.where(np.abs(top_10_data['z_score']) > 2, True, False)  # Consider Z-scores greater than 2 as outliers
```
